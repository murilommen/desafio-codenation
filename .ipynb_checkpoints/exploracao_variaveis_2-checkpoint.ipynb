```python
x_raw = df[['SG_UF_PROVA','TP_SEXO', 'TP_COR_RACA','TP_ST_CONCLUSAO','TP_ANO_CONCLUIU',
           'TP_ESCOLA','IN_TREINEIRO','IN_BAIXA_VISAO','IN_DISLEXIA','IN_DISCALCULIA',
          'IN_CEGUEIRA', 'IN_SURDEZ','IN_SABATISTA','IN_GESTANTE','IN_IDOSO',
          'TP_PRESENCA_CN','TP_PRESENCA_CH','TP_PRESENCA_LC','TP_PRESENCA_MT',
          'NU_NOTA_CN','NU_NOTA_CH','NU_NOTA_LC','NU_NOTA_REDACAO']]
y_raw = df['NU_NOTA_MT']
```

```python
y = df2.iloc[:,-1]
y.head()
```

```python
X = df2.iloc[:,:-1]
X.head()
```

```python
X = X.drop(['TP_SEXO_M'], axis='columns')
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.15,random_state=None)
```

```python
from sklearn.linear_model import LinearRegression
modelo = LinearRegression()
modelo.fit(X_train,y_train)
```

```python
y_previsto = modelo.predict(X_test)
```

```python
import statsmodels.api as sm 

X_train_const = sm.add_constant(X_train)
modelo_stats = sm.OLS(y_train, X_train, hasconst=True).fit()
```
